# Day 4 - Conversion & Campaign Performance Analysis

## Objective

Analyze relationships between campaign variables and customer conversions to identify patterns that can improve marketing ROI.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data/digital_marketing_campaign_dataset.csv")
```

```python
conversion_by_channel = df.groupby('CampaignChannel')['Conversion'].mean().sort_values(ascending=False)

conversion_by_channel
```

```python
plt.figure(figsize=(8,5))

conversion_by_channel.plot(kind='bar')

plt.title("Average Conversion Rate by Campaign Channel")
plt.xlabel("Campaign Channel")
plt.ylabel("Average Conversion Rate")

plt.xticks(rotation=45)

plt.show()
```

```python
conversion_by_type = df.groupby('CampaignType')['Conversion'].mean().sort_values(ascending=False)

conversion_by_type
```

```python
plt.figure(figsize=(8,5))

conversion_by_type.plot(kind='bar')

plt.title("Average Conversion Rate by Campaign Type")
plt.xlabel("Campaign Type")
plt.ylabel("Average Conversion Rate")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['AdSpend'], df['ConversionRate'])

plt.title("Ad Spend vs Conversion Rate")
plt.xlabel("Ad Spend")
plt.ylabel("Conversion Rate")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['Age'], df['ConversionRate'])

plt.title("Age vs Conversion Rate")
plt.xlabel("Age")
plt.ylabel("Conversion Rate")

plt.show()
```

# Day 4 Summary

## Campaign Performance Analysis

### Key Findings

- Conversion rates are similar across all marketing channels.
- Conversion campaigns achieve the highest average conversion rate.
- No strong relationship is observed between advertising spend and conversion rate.
- Customer age does not show a clear influence on conversion rate.

## Business Insights

- Marketing performance should not be evaluated using a single metric.
- ROI analysis should combine campaign performance, advertising spend, and customer behavior.
- Additional analyses on advertising platforms and customer engagement may provide deeper insights.
