# Day 3 - Customer Demographics & Campaign Analysis

## Objective

Explore customer demographics and marketing campaign characteristics to identify business insights that support marketing attribution and ROI analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data/digital_marketing_campaign_dataset.csv")

df.head()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Age'], bins=10)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Income'], bins=10)

plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Number of Customers")

plt.show()
```

```python
plt.figure(figsize=(6,4))

df['Gender'].value_counts().plot(kind='bar')

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

df['CampaignChannel'].value_counts().plot(kind='barh')

plt.title("Campaign Channel Distribution")
plt.xlabel("Number of Customers")
plt.ylabel("Campaign Channel")

plt.show()
```

# Day 3 Summary

## Customer Demographics Analysis

### Key Findings

- Customers are distributed across the age range of 18–69 years.
- Income levels vary from approximately 20,000 to 150,000, representing a diverse customer base.
- Female customers are more represented than male customers.

## Campaign Analysis

- Referral is the most frequently used marketing channel.
- PPC is the second most common marketing channel.
- Social Media has the lowest number of customer records.
- Campaign channels are fairly balanced, enabling effective comparison in later analyses.

## Business Insights

- Marketing campaigns target customers from different age and income groups.
- The dataset supports segmentation analysis based on demographics and campaign channels.
